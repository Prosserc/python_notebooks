```python
import pandas as pd
from pyproj import Transformer
```

```python
df = pd.read_excel('Schroot XY.xlsx')
df
```

```python
def convert(row, from_epsg='epsg:23031', to_epsg='epsg:4326'):
    t = Transformer.from_crs(crs_from=from_epsg, crs_to=to_epsg)
    return t.transform(row['X'], row['Y'])

df['latlon'] = df.apply(convert, axis=1)
df
```

```python
# split into lat & lon
df['long'] = df['latlon'].apply(lambda x: x[0])
df['lat'] = df['latlon'].apply(lambda x: x[1])
df.drop(columns=['latlon'], inplace=True)
df
```

```python
df.to_excel('.\Schroot_LatLon.xlsx')
```
