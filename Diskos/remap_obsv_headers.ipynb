```python
from diskos.reader import build_reader
import diskos.spec as spec
from diskos.sender_specific.merge import AptAnalId, ANAL_ID_COL
from diskos.spec import BlockType
import pandas as pd
```

```python
reader = build_reader(r'c:\data\diskos\6406_12-3_S__ROCK_AND_CORE__GCH_RAW_1.ASC', quiet_mode=True)
spec.SUPPRESS_WARNINGS = True  # some rows have missing cells - we know
parsed_file =reader.parse_file()
obsv_blocks = (b for b in parsed_file.blocks if b.block_type == BlockType.Observations)
```

```python
df = next(obsv_blocks).as_dataframe
df.head()
```

```python
def apply(original_df):
    df = original_df.copy()
    anal_codes = set([AptAnalId(id).anal_code for id in df[ANAL_ID_COL]])
    if len(anal_codes) == 1:
        code = anal_codes.pop()
        df.columns = [f"{col}.{code}" for col in df.columns]
    return df
```

```python
get_next_mod_df = lambda: apply(next(obsv_blocks).as_dataframe)
```
