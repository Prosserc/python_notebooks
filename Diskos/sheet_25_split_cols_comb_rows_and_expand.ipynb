```python
from diskos.reader import build_reader
import diskos.spec as spec
from diskos.sender_specific.merge import AptAnalId, ANAL_ID_COL, IGI_ID
```

```python
reader = build_reader(r'c:\data\diskos\6406_12-3_S__ROCK_AND_CORE__GCH_RAW_1.ASC', quiet_mode=True)
spec.SUPPRESS_WARNINGS = True  # some rows have missing cells - we know
parsed_file =reader.parse_file()
block_25 = next((b for b in parsed_file.blocks if b.block_def.block_no == '25'))
orig = block_25.as_dataframe
orig
```

```python
df = orig.copy()
cols_v1_v2 = {}
for col in df.columns:
    matches = [k for k in cols_v1_v2.keys() if col.startswith(k)]
    if matches:
        cols_v1_v2[matches[0]] = col
    else:
        cols_v1_v2[col] = None
cols_v1_v2
```

```python
df1 = df[cols_v1_v2.keys()]
df1
```

```python
for k, v in cols_v1_v2.items():
    if not v:
        cols_v1_v2[k] = f"{k}_2"
        df[f"{k}_2"] = ""
df2 = df[cols_v1_v2.values()]
#df2.dropna(subset=["AnalID_2"])
df2 = df2[df2["AnalID_2"] != ""]
df2
```
